```python
# Create the llm_config
gpt4_config = {
    "seed": 42,
    "temperature": 0,
    "config_list": config_list,
    "timeout": 60,
}
```

```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager

# Create the user proxy
user_proxy = UserProxyAgent(
    name = "Admin",
    system_message = "A human admin. Interact with the manager to get the finalised job description. This finalised job description should be approved by this admin.",
    code_execution_config = False,
)
# Create the drafter
drafter = AssistantAgent(
    name = "Drafter",
    llm_config = gpt4_config,
    system_message = "Drafter. You write a well structured and detailed Job Description based on the Job Name received by you. You follow the proper industry standards and keywords in the job description.",
)
# Create a critic
critic = AssistantAgent(
    name = "Critic",
    llm_config = gpt4_config,
    system_message = "Critic. You crawl through the web to check for job description mentioned against similar or same Job Name received by you. You generally check the Job Postings of reputed companies. You rate the job description created by 'Drafter' agent and create a better job description as per the market standards. You also give feedback.",
)
# Create a manager
manager = AssistantAgent(
    name = "Manager",
    llm_config = gpt4_config,
    system_message = "You are the 'Manager' in charge of coordinating job description refinement between the 'Critic' and 'Drafter' agents. Your role is to ensure the final job description is both accurate and appealing to potential candidates. The 'Drafter' generates an initial job description based on the job name provided by the User Proxy, and the 'Critic' matches it against industry-standard job postings. Please facilitate effective communication between 'Drafter' and 'Critic' to refine the job description and provide the best output for the User Proxy.",
)

groupchat = GroupChat(agents=[user_proxy, drafter, critic, manager], messages=[], max_round=50)
manager = GroupChatManager(groupchat=groupchat, llm_config=gpt4_config)
```

## Start Chat

```python
user_proxy.initiate_chat(
    manager,
    message="""
    Job Name: Software Engineer 1 - Payments System
    """,
)
```
